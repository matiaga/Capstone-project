#### 3. Write a to find styles that have more than one color.

```python
query = '''
    WITH temp AS (SELECT product_code, color_1 AS color FROM Table2 WHERE color_1 IS NOT NULL
    
    UNION
        SELECT product_code, color_2 AS color FROM Table2 WHERE color_2 IS NOT NULL
    
    UNION
        SELECT product_code, color_3 AS color  FROM Table2 WHERE color_3 IS NOT NULL
    
    UNION
        SELECT product_code, color_4 AS color  FROM Table2 WHERE color_4 IS NOT NULL
    
    UNION
        SELECT product_code, other_color AS color  FROM Table2 WHERE other_color IS NOT NULL)
       
    SELECT product_code, 
        COUNT(DISTINCT color) AS num_colors
    FROM temp 
    WHERE product_code IS NOT NULL
    GROUP BY product_code
    HAVING num_colors > 1
    ORDER BY num_colors DESC;
'''
# Fetch data from database
results = cursor.execute(query)
rows = results.fetchall()

for i in rows:
    print(i)

print(f'\nTotal records recovered: {len(rows)}')
```

#### 4  Write a query to find the count of sizes available for each color for a specific style code.


```python
query = '''
    WITH temp AS (SELECT product_code, count_of_sizes, color_1 AS color FROM Table2 WHERE color_1 IS NOT NULL
    
    UNION
        SELECT product_code, count_of_sizes, color_2 AS color FROM Table2 WHERE color_2 IS NOT NULL
    
    UNION
        SELECT product_code, count_of_sizes, color_3 AS color  FROM Table2 WHERE color_3 IS NOT NULL
    
    UNION
        SELECT product_code, count_of_sizes, color_4 AS color  FROM Table2 WHERE color_4 IS NOT NULL
    
    UNION
        SELECT product_code, count_of_sizes, other_color AS color  FROM Table2 WHERE other_color IS NOT NULL)
       
    SELECT product_code, 
        color,
        SUM(count_of_sizes) AS count_of_sizes
    FROM temp 
    WHERE product_code IS NOT NULL
    GROUP BY product_code, color
    ORDER BY count_of_sizes DESC;
'''

# Fetch data from database
results = cursor.execute(query)
rows = results.fetchall()

for i in rows:
    print(i)

print(f'\nTotal records recovered: {len(rows)}')
```

#### 1. Write a query calculates the average comfort rating for a specific product based on its reviews.
- Only products that have at least one review are considered to have comfort or star rating metrics.
- You can enter a snippet of the code and it will show all the products that match

```python
product = input("Please, enter the product code:") 
print(f"Searching for: {product}")

query = f'''
    SELECT t3.product_code, 
        SUM(t3.n_reviews) AS total_reviews, 
        AVG(t3.comfort) AS avg_comfort, 
        AVG(t3.star_rating) AS avg_rating
    FROM Table3 t3
    WHERE t3.product_code LIKE '{product}%'
        AND (t3.n_reviews IS NOT NULL AND t3.n_reviews > 0)
    GROUP BY t3.product_code
    ORDER BY t3.star_rating DESC;
'''
print(query)
# Fetch data from database
results = cursor.execute(query)
rows = results.fetchall()

for i in rows:
    print(f"{i}")

print(f'\nTotal records recovered: {len(rows)}')
```

#### 6. Write a  query that  identifies products with a star rating above the average star rating for their respective size.


```python
query = '''
    SELECT t1.ShoeName, t1.product_code, t3.star_rating, t2.count_of_sizes
    FROM Table1 t1
    JOIN Table2 t2 ON t1.product_code = t2.product_code
    JOIN Table3 t3 ON t1.product_code = t3.product_code
    WHERE t3.star_rating > (
        SELECT AVG(t3_avg.star_rating)
        FROM Table2 t2_avg
        JOIN Table3 t3_avg ON t2_avg.product_code = t3_avg.product_code
        WHERE t2_avg.count_of_sizes = t2.count_of_sizes )
    GROUP BY t1.ShoeName, t1.product_code, t3.star_rating, t2.count_of_sizes
    ORDER BY t2.count_of_sizes DESC, t3.star_rating DESC;
'''

# Fetch data from database
results = cursor.execute(query)
rows = results.fetchall()

for i in rows:
    print(i)

print(f'\nTotal records recovered: {len(rows)}')
```

#### 7. Write  a  query finds products with the highest comfort rating in each category.

```python
query = '''
    WITH temp AS (
    SELECT
        t1.ShoeName,
        t1.Category,
        t3.comfort,
        t3.product_code,
        ROW_NUMBER() OVER (PARTITION BY t1.Category ORDER BY t3.comfort DESC) AS rn
    FROM
        Table1 t1
        INNER JOIN Table3 t3 ON t1.product_code = t3.product_code
    )
    SELECT
        ShoeName,
        Category,
        comfort,
        product_code
    FROM
        temp
    WHERE
        rn = 1;

'''

# Fetch data from database
results = cursor.execute(query)
rows = results.fetchall()

for i in rows:
    print(i)

print(f'\nTotal records recovered: {len(rows)}')
```
